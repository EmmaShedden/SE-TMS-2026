## Looking at the dataset

```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
filepath = 'preprocessed-data/full-data-version-9/programming-full-data.csv'
da = pd.read_csv(filepath)
da.head(2)
```

```python
plt.hist(da.loc[da['problem_type'] == 'short', 'duration'], bins=20, alpha=0.5, label='FITB')
# plt.xscale('log')
plt.hist(da.loc[da['problem_type'] == 'long', 'duration'], bins=20, alpha=0.5, label='LR')
plt.xlabel('Duration (seconds?)')
plt.ylabel('Observations')
plt.title('Distribution of Problem Durations')
plt.legend()
plt.show()
```

6. `TMS_Type` (CM, WM): treatment condition

8. `problem_id`: string, unique question identifier
