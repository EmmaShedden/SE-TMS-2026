# Script to redact PID from raw and preprocessed data files

Only redacted files (not originals) should ever be commited to the public repo.

Files touched:
- `raw-data/demographics.csv`
- `raw-data/pain_reports.csv`
- `raw-data/participant_tracking.csv`
- `preprocessed-data/full-data-version-9/nback-full-data.csv`
- `preprocessed-data/full-data-version-9/programming-full-data.csv`

```python
import numpy as np
import pandas as pd
```

## Raw data > demographics

```python
df = pd.read_csv('raw-data/demographics.csv')
df = df.sample(frac=1).reset_index(drop=True)
df = df.sort_values(by=['age', 'programming_exp']).reset_index(drop=True)
```

```python
for col in ['participant_id', 'background_2', 'occupation', 'current_school_year', 'disability_2']:
    df.loc[~df[col].isna(), col] = 'REDACTED'
```

```python
df.to_csv('raw-data/participant_tracking.csv', index=False, na_rep='')
```

## Raw data > pain reports

```python
df = pd.read_csv('raw-data/pain_reports.csv')
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.loc[~df['participant_id'].isna(), 'participant_id'] = 'REDACTED'
```

```python
df.to_csv('raw-data/pain_reports.csv', index=False, na_rep='')
```

## Raw data > participant tracking

```python
df = pd.read_csv('raw-data/Participant Tracking 2024_2025 - Participant Schedule.csv')
df.loc[~df['Email'].isna(), 'Email'] = 'REDACTED'
df.loc[~df['ID'].isna(), 'ID'] = 'REDACTED'
```

```python
# Initial contact column: mm/dd/yyyy to mm/yyyy
# Session date columns to days after contacted
# (formatting) Session times to 24 hour format
```

```python
df.to_csv('raw-data/participant_tracking.csv', index=False, na_rep='')
```

## Preprocessed n-back data

```python
df = pd.read_csv('preprocessed-data/full-data-version-9/nback-full-data.csv')
```

```python
redact_cols = [
    'start_time_real',
    #'start_time_experiment',
    #'end_time_experiment',
    'age',
    'gender',
    'background',
    'background_2',
    'english_speaker',
    'occupation',
    'disability',
    'current_school_year',
    'programming_exp',
    'disability2'
]

for col in redact_cols:
    df[col] = 'REDACTED'
```

```python
ids = df['participant_id'].unique()
ids = np.random.permutation(ids)
id_map = {real_id: f'REDACTED_{i}' for i, real_id in enumerate(ids)}
df['participant_id'] = df['participant_id'].map(id_map)
df = df.sort_values(by=['participant_id', 'session', 'nback_type', 'problem_order']).reset_index(drop=True)
```

```python
df.head()
```

```python
df.to_csv('preprocessed-data/full-data-version-9/nback-full-data.csv', index=False, na_rep='')
```

## Preprocessed programming data

```python
df = pd.read_csv('preprocessed-data/full-data-version-9/programming-full-data.csv')
```

```python
for col in redact_cols:
    df[col] = 'REDACTED'
```
